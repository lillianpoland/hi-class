## The First Four Moments: Mean, Variance, Skew, and Kurtosis

When we talk about a probablity dist, we often care about at four things:
- The expected value, or where a distribution is centered (mean)
- The amount of variation from that centered value (variation)
- Whether or not our distribution favors, leans towards, or skews towards certain values. (skew)
- How often to expect unusually high or low values. (kurtosis)

### Playing with Beta

**Task1**:

Plot a [preliZ Beta distribution](https://preliz.readthedocs.io/en/latest/api_reference.html#preliz.distributions.beta.Beta) with $α=1$ and $β=1$.

````python
p = pz.Beta(alpha=1,beta=1)
p.plot_pdf()
````

**Task2**:

Find the mean, variance, skew, and kurtosis of a $Beta(1,1)$ distribution. This is the same as asking for the first four moments of a $Beta(1,1)$ distribution.

*Hint*: preliZ has built in function that does this for you.

````python
mean, variance, skew, kurtosis = p.moments()
print(f"Mean: {mean}")
print(f"Variance: {variance}")
print(f"Skew: {skew}")
print(f"Kurtosis: {kurtosis}")
````

**Task3**:

Plot the following distributions: $Beta(1,1)$, $Beta(0.9,0.9)$, $Beta(2,1)$, $Beta(1,2)$, $Beta(7,11)$, $Beta(96,42)$, $Beta(5,5)$, $Beta(500,500)$, $Beta(0.8,0.9)$.

Then, determine which of those distribtuions has:
- the largest/smallest mean
- the largest/smallest variance
- the largest/smallest skew
- the largest/smallest kurtosis

````python
p1 = pz.Beta(alpha=1,beta=1)
p2 = pz.Beta(alpha=0.9,beta=0.9)
p3 = pz.Beta(alpha=2,beta=1)
p4 = pz.Beta(alpha=1,beta=2)
p5 = pz.Beta(alpha=7,beta=11)
p6 = pz.Beta(alpha=96,beta=42)
p7 = pz.Beta(alpha=5,beta=5)
p8 = pz.Beta(alpha=500,beta=500)
p9 = pz.Beta(alpha=0.8,beta=0.9)

p1.plot_pdf()
p2.plot_pdf()
p3.plot_pdf()
p4.plot_pdf()
p5.plot_pdf()
p6.plot_pdf()
p7.plot_pdf()
p8.plot_pdf()
p9.plot_pdf()

print(p1.mean(), p2.mean(), p3.mean(), p4.mean(), p5.mean(), p6.mean(), p7.mean(), p8.mean(), p9.mean())
print(p1.var(), p2.var(), p3.var(), p4.var(), p5.var(), p6.var(), p7.var(), p8.var(), p9.var())
print(p1.skewness(), p2.skewness(), p3.skewness(), p4.skewness(), p5.skewness(), p6.skewness(), p7.skewness(), p8.skewness(), p9.skewness())
print(p1.kurtosis(), p2.kurtosis(), p3.kurtosis(), p4.kurtosis(), p5.kurtosis(), p6.kurtosis(), p7.kurtosis(), p8.kurtosis(), p9.kurtosis())
````

The largest mean is Beta(96,42). The smallest mean is Beta(1,2).

The largest variance is Beta (0.8,0.9). The smallest variance is Beta(500,500)

The largest skewness is Beta(1,2). The smallest skewness is Beta(2,1)

The largest kurtosis is Beta(500,500). The smallest kurtosis is (0.8,0.9)
